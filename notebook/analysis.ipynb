## Trailer Activity Report

### Analyzing the data in the Open Order Report:

```python
trailer_report = pd.read_csv('/Users/jamesferry/Documents/GitHub/BradshawDwellTimeDash/data/raw/WM-BRD-YRD101 Trailer Activity Report (33).csv')
```

```python
trailer_report.columns
```

```python
columns_to_keep = ['CHECKIN DATE TIME', 'APPOINTMENT DATE TIME', 'CHECKOUT DATE TIME', 'CARRIER', 'VISIT TYPE', 'ACTIVITY TYPE ', 'Date/Time', 'SHIPMENT_ID']
trailer_report = trailer_report.drop(columns=trailer_report.columns.difference(columns_to_keep))

trailer_report.head()
```

```python
trailer_report= trailer_report[trailer_report['ACTIVITY TYPE '] == 'CLOSED']
trailer_report.head()
```

```python
unique_visit = trailer_report['VISIT TYPE'].unique()
print(unique_visit)
```

```python
trailer_report = trailer_report[trailer_report['VISIT TYPE'].isin(['Pickup Load', 'Live Load'])]
trailer_report.head()
```

```python
trailer_report.isnull().sum()
```

```python
trailer_report['SHIPMENT_ID'] = trailer_report['SHIPMENT_ID'].replace('', np.nan)
trailer_report = trailer_report.dropna(subset = ['SHIPMENT_ID'])

trailer_report.isnull().sum()
```

```python
dupes = trailer_report.apply(lambda col: col.duplicated().sum())
print(dupes)
```

```python
trailer_report = trailer_report.sort_values(by=['Date/Time', 'SHIPMENT_ID'], ascending=[False, True])
```

```python
trailer_report = trailer_report.drop_duplicates(subset='SHIPMENT_ID', keep='first')

dupes = trailer_report.apply(lambda col: col.duplicated().sum())
print(dupes)
```

```python
trailer_report.dtypes
```

```python
trailer_report['CHECKIN DATE TIME'] = pd.to_datetime(trailer_report['CHECKIN DATE TIME'])
trailer_report['APPOINTMENT DATE TIME'] = pd.to_datetime(trailer_report['APPOINTMENT DATE TIME'])
trailer_report['CHECKOUT DATE TIME'] = pd.to_datetime(trailer_report['CHECKOUT DATE TIME'])
trailer_report['VISIT TYPE'] = trailer_report['VISIT TYPE'].astype('category')
trailer_report['ACTIVITY TYPE '] = trailer_report['ACTIVITY TYPE '].astype('category')
trailer_report['Date/Time'] = pd.to_datetime(trailer_report['Date/Time'])
trailer_report['SHIPMENT_ID'] = trailer_report['SHIPMENT_ID'].astype('int64')
```

```python
def required_time(row):
    if row['VISIT TYPE'] == 'Live Load':
        return row['APPOINTMENT DATE TIME'] + pd.Timedelta(minutes = 15)
    else:
        return row['APPOINTMENT DATE TIME'] + pd.Timedelta(hours = 24)

trailer_report['Required Time'] = trailer_report.apply(required_time, axis=1)

trailer_report.head()
```

```python
def compliance(row):
    if row['Required Time'] >= row['CHECKIN DATE TIME']:
        return 'On Time'
    else:
        return 'Late'
    
trailer_report['Compliance'] = trailer_report.apply(compliance, axis=1)

trailer_report.head()
```

```python
trailer_report['Scheduled Date'] = trailer_report['APPOINTMENT DATE TIME'].dt.date
trailer_report.head()
```

```python
trailer_report['Week'] = trailer_report['APPOINTMENT DATE TIME'].dt.isocalendar().week
trailer_report.head()
```

```python
trailer_report['Month'] = trailer_report['APPOINTMENT DATE TIME'].dt.month
trailer_report.head()
```

```python
trailer_report['APPOINTMENT DATE TIME'] = trailer_report['APPOINTMENT DATE TIME'].replace('', np.nan)
trailer_report = trailer_report.dropna(subset = ['APPOINTMENT DATE TIME'])

trailer_report['Week'] = trailer_report['Week'].astype('int64')
trailer_report['Month'] = trailer_report['Month'].astype('int64')
```

```python
trailer_report.isnull().sum()
```

```python
trailer_report.columns
```

```python
columns_to_keep = ['CHECKIN DATE TIME', 'CHECKOUT DATE TIME', 'CARRIER', 'VISIT TYPE', 'Date/Time', 'SHIPMENT_ID', 'Required Time', 'Compliance', 'Scheduled Date', 'Week', 'Month']
trailer_report = trailer_report.drop(columns=trailer_report.columns.difference(columns_to_keep))

trailer_report.head()
```

```python
trailer_report.rename(columns={'CHECKIN DATE TIME': 'Checkin DateTime', 'CHECKOUT DATE TIME': 'Checkout DateTime', 'CARRIER': 'Carrier', 'VISIT TYPE': 'Visit Type', 'Date/Time': 'Loaded DateTime', 'SHIPMENT_ID': 'Shipment ID'}, inplace=True)
trailer_report.columns
```

```python
trailer_report['Scheduled Date'] = pd.to_datetime(trailer_report['Scheduled Date'])
```

```python
trailer_report.dtypes
```

```python
trailer_report['Checkout DateTime'] = trailer_report['Checkout DateTime'].replace('', np.nan)
trailer_report = trailer_report.dropna(subset = ['Checkout DateTime'])
```

```python
trailer_report.isnull().sum()
```

```python
trailer_report.to_csv('/Users/jamesferry/Documents/GitHub/BradshawDwellTimeDash/data/processed/trailer_report.csv', index = False)
```

## Create a Merged Dataset Using Duck DB

```python
import duckdb

con = duckdb.connect(database = ':memory:')

con.execute("CREATE TABLE open_dock AS SELECT * FROM read_csv_auto('/Users/jamesferry/Documents/GitHub/BradshawDwellTimeDash/data/processed/open_dock.csv')")
con.execute("CREATE TABLE open_order AS SELECT * FROM read_csv_auto('/Users/jamesferry/Documents/GitHub/BradshawDwellTimeDash/data/processed/open_order.csv')")
con.execute("CREATE TABLE trailer_report AS SELECT * FROM read_csv_auto('/Users/jamesferry/Documents/GitHub/BradshawDwellTimeDash/data/processed/trailer_report.csv')")

print(con.execute("SELECT * FROM open_dock LIMIT 5").fetchdf())
print(con.execute("SELECT * FROM open_order LIMIT 5").fetchdf())
print(con.execute("SELECT * FROM trailer_report LIMIT 5").fetchdf())
```

```python
merged_df = con.execute("""
    SELECT
        open_dock."SO Number" AS "Dock SO Number",
        open_dock."Dwell Time",
        open_order."SO Number" AS "Order SO Number",
        open_order."Appt DateTime",
        open_order."Shipment ID"
    FROM open_dock
    LEFT JOIN open_order
    ON POSITION(CAST(open_order."SO Number" AS VARCHAR) IN CAST(open_dock."SO Number" AS VARCHAR)) > 0
""").fetchdf()

# Check if merged_df is created correctly
if merged_df is not None:
    print(merged_df.head())
else:
    print("Query did not return any data.")
```

```python
merged_df.head()
```

```python
merged_df.columns
```

```python
columns_to_keep = ['Dock SO Number', 'Dwell Time', 'Appt DateTime', 'Shipment ID']
merged_df = merged_df.drop(columns=merged_df.columns.difference(columns_to_keep))

merged_df.rename(columns={'Dock SO Number': 'SO Number'}, inplace=True)
```

```python
merged_df.head()
```

```python
merged_df['SO Number'] = merged_df['SO Number'].astype('object')
```

```python
merged_df.dtypes
```

```python
if merged_df is not None:
    dwell_and_ontime_compliance = con.execute("""
        SELECT 
            trailer_report."Shipment ID",
            merged_df."SO Number",
            merged_df."Appt DateTime",
            trailer_report."Required Time",
            trailer_report."Checkin DateTime",
            trailer_report."Checkout DateTime",
            trailer_report.Carrier,
            trailer_report."Visit Type",
            trailer_report."Loaded DateTime",
            trailer_report.Compliance,
            merged_df."Dwell Time",
            trailer_report."Scheduled Date",
            trailer_report.Week,
            trailer_report.Month
        FROM trailer_report
        INNER JOIN merged_df ON trailer_report."Shipment ID" = merged_df."Shipment ID"
    """).fetchdf()

    # Check if dwell_and_ontime_compliance is created correctly
    if dwell_and_ontime_compliance is not None:
        dwell_and_ontime_compliance.head()
    else:
        print("Query did not return any data.")
else:
    print("merged_df is not created correctly.")
```

```python
dwell_and_ontime_compliance = dwell_and_ontime_compliance.drop_duplicates(subset='Shipment ID', keep='first')
```
